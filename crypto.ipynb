```python
import requests
import json
from datetime import datetime, timedelta
import calendar

# Define the API key
API_KEY = "473b3b0f-eeeb-4b3d-875b-465d191fb16a"

# Define the symbols for each asset in USD.
# Adjust these as needed based on the exchanges available.
symbols = {
    "BTC": "BITSTAMP_SPOT_BTC_USD",
    "ETH": "BITSTAMP_SPOT_ETH_USD",
    "DOGE": "BINANCE_SPOT_DOGE_USDT", 
    "XRP": "BITSTAMP_SPOT_XRP_USD"
}

# CoinAPI OHLCV endpoint base URL for historical data
base_url = "https://rest.coinapi.io/v1/ohlcv"

# Headers with the API key
headers = {
    "X-CoinAPI-Key": API_KEY
}

# Function to get the start and end ISO8601 strings for a given month and year.
def get_month_range(year, month):
    start = datetime(year, month, 1)
    # Determine last day of the month.
    last_day = calendar.monthrange(year, month)[1]
    # The end time is the start of the next month.
    if month == 12:
        end = datetime(year + 1, 1, 1)
    else:
        end = datetime(year, month + 1, 1)
    return start.isoformat(), end.isoformat()

# List the years you want to query
years = [2022, 2023, 2024]

# Create a container to store results
all_data = {}

# Loop through each asset, year, and month
for asset, symbol in symbols.items():
    all_data[asset] = []
    for year in years:
        for month in range(1, 13):
            time_start, time_end = get_month_range(year, month)
            
            # Construct the endpoint URL for the asset symbol.
            url = f"{base_url}/{symbol}/history"
            
            # Set parameters for monthly data (using "1MTH" period)
            params = {
                "period_id": "1MTH",   # monthly candles
                "time_start": time_start,
                "time_end": time_end,
                "limit": 1             # Expect one candle per month
            }
            
            print(f"Requesting {asset} data for {year}-{month:02d}...")
            response = requests.get(url, headers=headers, params=params)
            
            if response.status_code == 200:
                data = response.json()
                # CoinAPI returns a list of candles. If data exists, use the first candle.
                if data:
                    candle = data[0]
                    # Add asset symbol and the corresponding month/year for clarity.
                    candle["asset_symbol"] = asset
                    candle["year"] = year
                    candle["month"] = month
                    all_data[asset].append(candle)
                else:
                    print(f"No data returned for {asset} in {year}-{month:02d}.")
            else:
                print(f"Error for {asset} {year}-{month:02d}: {response.status_code} - {response.text}")

# Print out the collected data in JSON format (or process further as needed)
print(json.dumps(all_data, indent=4))
```

```python
import pandas as pd
import json

# Example: Assume your JSON output is stored in a variable called all_data.

# Flatten the dictionary into a list of records.
records = []
for asset, data_list in all_data.items():
    # Each record already includes an "asset_symbol" key, but this ensures all records are collected.
    records.extend(data_list)

# Create the DataFrame
df = pd.DataFrame(records)

# Optional: Convert date/time strings to datetime objects for better handling
df['time_period_start'] = pd.to_datetime(df['time_period_start'])
df['time_period_end'] = pd.to_datetime(df['time_period_end'])
df['time_open'] = pd.to_datetime(df['time_open'])
df['time_close'] = pd.to_datetime(df['time_close'])
```
